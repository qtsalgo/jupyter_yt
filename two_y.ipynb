# How to plot two Y axis, matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('fdax_2y.csv')
```

```python
df.head()
```

```python
x = df['bar']
y1 = df['close']
y2 = df['cdf']

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

curve1 = ax1.plot(x, y1, label='close',color='orange')
curve2 = ax2.plot(x, y2, label='cdf',color='grey')

plt.plot()
plt.show()
```
