## Make sure you have imported Pandas and Seaborn


```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("white");
```

## Create a data frame from your dataset 

```python
df = pd.read_csv('detail_markov_probabilities.csv')
df.shape
```

## sns.relplot - This function shows the relationship between two variables

- We will look at two volitility variables that should create a smile curve

- First distribution includes hue




```python
sns.relplot(x="IVS", y="IVK",hue='Pnl', data=df);

# remove hue
sns.relplot(x="IVK", y="IVS", data=df);
```

## View the variables distribution with skew and kurtosis

```python
a = sns.displot(data=df, x="IVK",kind ='kde',color='darkred')
b = sns.displot(data=df, x="IVS",kind ='kde',color='darkBLUE')


for ax in a.axes.ravel():
    ax.text(x=0.97, y=0.97, transform=ax.transAxes, s="Skewness: %f" % df['IVK'].skew(),\
        fontsize=10, verticalalignment='top', horizontalalignment='right')
    ax.text(x=0.97, y=0.91, transform=ax.transAxes, s="Kurtosis: %f" % df['IVK'].kurt(),\
        fontsize=10, verticalalignment='top', horizontalalignment='right')
for ax in b.axes.ravel():    
    ax.text(x=0.97, y=0.97, transform=ax.transAxes, s="Skewness: %f" % df['IVS'].skew(),\
        fontsize=10, verticalalignment='bottom', horizontalalignment='right')
    ax.text(x=0.97, y=0.91, transform=ax.transAxes, s="Kurtosis: %f" % df['IVS'].kurt(),\
        fontsize=10, verticalalignment='bottom', horizontalalignment='right')

    
```
