# Matplotlib Line

```python
import matplotlib.pyplot as plt
import numpy as np;
```

### Linestyle

You can use the keyword argument linestyle, or just ls, to change the style of the plotted line:

```python
ypoints = np.array([100, 110, 103, 113, 109, 123])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'solid')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashdot')
plt.show()
```

## Line Color

You can use the keyword argument colour or the shorter c to set the line colour:

```python
plt.plot(ypoints, color = 'r')
plt.show() 
```

```python
plt.plot(ypoints, color = 'b')
plt.show() 
```

```python
plt.plot(ypoints, color = 'g')
plt.show() 
```

### Line Width

You can use the keyword argument line width or the shorter lw to change the width of the line.


```python
ypoints = np.array([100, 110, 103, 113, 109, 123])

plt.plot(ypoints, linewidth = '10')
plt.show()
```

```python
plt.plot(ypoints, linewidth = '20')
plt.show()
```

## Multiple Lines

You can plot as many lines as you like by adding more plt.plot() functions:

```python
y1 = np.array([100, 110, 103, 113, 109, 123])
y2 = np.array([110, 103, 113, 109, 123, 135])

plt.plot(y1)
plt.plot(y2)

plt.show()
```
