# Create a  BTC/ETH pair and plot with matplot.lib / seaborn

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete");
```

```python


# Path of the file to read
df = pd.read_csv('3_btc_eth_pair.csv')
```

# Check data frame head and shape

We have created a daily data set for BTC_ETH

- Where: BTC_ETH = btc_close / eth_close

```python
df.head()
```

```python
df.shape
```

# Plot the BTC_ETH price

```python
# Set width / height of the figure
plt.figure(figsize=(20,6))

# Add a title
plt.title("BTC_ETH Pair")
sns.lineplot(data=df['BTC_ETH'], label="Pair Daily Close Price")

# Add a label for the x axis
plt.xlabel("Bars")
```

# Plot the underlying coins, BTC and ETH, on seperate charts

```python
# Set width / height of the figure
plt.figure(figsize=(20,6))

# Add a title
plt.title("BTC")
sns.lineplot(data=df['btc_close'], label="BTC Daily Close Price")

# Add a label for the x axis
plt.xlabel("Bars")

# Set width / height of the figure
plt.figure(figsize=(20,6))

# Add a title
plt.title("ETH")
sns.lineplot(data=df['eth_close'], label="ETH Daily Close Price")

# Add a label for the x axis
plt.xlabel("Bars")
```
