# Continuous variable plots using matplotlib & seaborn

#### import pandas matplotlib and seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv('dailyfdaxRaw.csv')
```

```python
data.head()
```

## Scatter Plot

#### Scatterplot showing high values in relation to the low values

```python
plt.title('high v low')
sns.scatterplot(data['High'],data['Low'])
plt.show()
```

## Histogram

#### Histogram of 'Close prices into 10 bins using Seaborn Displot 

```python
sns.displot(data['Close'],bins=10,kde=True)
plt.show()
```

## DataFrame Histogram

#### Histograms for all numerical columns in the data DataFrame using Pandas hist:

```python
 data.hist(figsize=(20,30),layout=(9,4));
```

## Density Plot

#### Density plot for the close price variable using Seaborn kdeplot:

```python

plt.figure(figsize=(10,5))
plt.xlabel('Density')
plt.ylabel('Close')
sns.kdeplot(data['Close'],shade=True)
plt.show()
```

## Box Plot

#### Box plot for the price variable using Seaborn and Matplotlib:

```python

plt.figure(figsize=(10,5))
plt.ylabel('Close')
sns.boxplot(y=data['Close'])
plt.show()
```
