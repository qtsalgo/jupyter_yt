# Select Rows by Multiple Conditions Using Pandas 

```python
import pandas as pd
```

```python
data = pd.read_csv('groupby.csv')
```

```python
data.head()
```

```python
data.shape
```

- Select Rows that Meet Multiple Conditions

```python
#select rows where D is equal to 8 and WIN is equal to 1
data.loc[((data['D'] == 8) & (data['Win'] == 1))]
```

# Select Rows that Meet One of Multiple Conditions

```python
#select rows where D is greater than 8 or D is less than 8
data.loc[((data['D'] > 8) | (data['D'] < 8))]
```
