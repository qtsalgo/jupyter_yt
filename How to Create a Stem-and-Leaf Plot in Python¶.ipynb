```python
import pandas as pd
import stemgraphic
```

```python
pnl1 = [2500,2500,500,-800,-900,-154,929,983,839,840,-2786,900,0,0,760,177,0,0,1000,0]
pnl2 = [2066,-3133,4682,2430,3478,1641,801,0,4600,2435,2012,1243,450,211,2709,3339,0,1325,-1200,0]
```

```python
#create stem-and-leaf plot
fig, ax = stemgraphic.stem_graphic(pnl2)
```

- Numbers in the red boxes are the min and max values of the data set
- Numbers in the left column display the aggregated count values, for example 20 at the top tells us there were a total of twenty samples inthe data set
- Numbers in the middle column show the stems
- Numbers in the right column are the leaves
