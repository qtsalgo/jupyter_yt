```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm;

```

# How to create the bell curve

```python
#create x-values from -4 to 4 in increments of .001
x = np.arange(-4, 4, 0.1)

#create y-values that correspond to normal pdf with mean=0 and sd=1 
y = norm.pdf(x,0.01)

#define plot 
fig, ax = plt.subplots(figsize=(20,13))
ax.plot(x,y)

#choose a plot style  
plt.style.use('fivethirtyeight')
plt.show()
```

# How to fill in area under the curve 

```python
#create x-values from -4 to 4 in increments of .001
x = np.arange(-4, 4, 0.1)

#create y-values that correspond to normal pdf with mean=0 and sd=1 
y = norm.pdf(x,0.01)

#define plot 
fig, ax = plt.subplots(figsize=(20,13))
ax.plot(x,y)#specify area under the curve to fill 
x_fill = np.arange(-2, 2, 0.001)
y_fill = norm.pdf(x_fill,0,1)
ax.fill_between(x_fill,y_fill,0, alpha=0.5, color='red')
```
